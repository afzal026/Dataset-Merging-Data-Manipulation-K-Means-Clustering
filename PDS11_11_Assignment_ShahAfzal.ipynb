```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv")
df
```

```python

df.fillna({"Age" : df["Age"].mean(), "Cabin": "Unknown"}, inplace = True)
df
```

```python
df['number of relatives'] = df['SibSp'] + df['Parch']
df
```

```python
dead_Pes = df[df['Survived']== 0]


plt.figure(figsize=(8,6))
plt.scatter(dead_Pes['number of relatives'], dead_Pes['Age'], color='red')
plt.xlabel('Number of Relatives')
plt.ylabel('Age')
plt.title('Scatter Plot of Dead Passengers: Number of Relatives vs Age')
plt.grid(True)
plt.show()
```
